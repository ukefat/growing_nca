```python
from nca.model import *
from nca.rollout import *
from nca.viz import *
from nca.stats import *

import torch
import numpy as np
import imageio
from IPython.display import Image, display
from IPython.display import Video
```

```python
device = torch.device("cpu")
model = NCA()
state = torch.load("../../snapshots/froggy_model_noregen.pth", map_location=device)
model.load_state_dict(state)
model.eval()  # VERY IMPORTANT
```

```python
x0 = seed(1)  # initial seed, shape [1, C, H, W]
with torch.no_grad():
    states = rollout(
        model,
        x0,
        steps=1500,
        ablate_channel=None,
        width_cut=None,
        killEye=150,
        prune=True,
        realEyeReplace=None,
        realEyeCapture=None,
        fakeEyeReplace=None,
        inhibit=None,
        injectionChannel=None,
        injectionTime=300,
        injectionConcoction=True
    )
# states: (T, C, H, W)
```

```python
frames = [frame_to_rgb_uint8(f[:4]) for f in states] # Select first 4 channels (RGBA) and convert

save_crisp_mp4(
    frames,
    output_file="output.mp4",
    fps=120,
    upscale=1,
    frames_dir="frames", 
    quiet=True
)

# local file
Video("output.mp4", embed=True, width=400, height=400)
```

```python
c_frames = [channel_to_rgb(f[12]) for f in states] # Select first 4 channels (RGBA) and convert

save_crisp_mp4(
    c_frames,
    output_file="output.mp4",
    fps=120,
    upscale=1,
    frames_dir="frames", 
    quiet=True
)

# local file
Video("output.mp4", embed=True, width=400, height=400)
```
